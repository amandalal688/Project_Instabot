```python
#importing all libraried
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
from selenium.common.exceptions import StaleElementReferenceException
from selenium.common.exceptions import ElementClickInterceptedException
from selenium.common.exceptions import NoSuchElementException
```

```python
#initializing the webdriver
#you can set your chrome driver executable path underneath
driver = webdriver.Chrome()
driver.maximize_window()
```

```python
#Initializing the request to the browser for instagram.com
driver.get('https://www.instagram.com/')
```

```python
#update your username and password for login
username = 'SAMPLE USERNAME'
password = 'SAMPLE PASSWORD'
```

```python
#initializing object for WebDriverWait
wait = WebDriverWait(driver , 10)
```

# Problem 1 - Login to your Instagram

```python
#this function lets you login in your Instagram handle
def LogIn(username, password):
    try :
        #locating the username textbox and sending username
        user_name = wait.until(EC.presence_of_element_located((By.NAME,"username")))
        #making sure that user name input box is clear
        user_name.clear()
        time.sleep(1)
        user_name.send_keys(username)
        #locating the password textbox and sending password
        pwd = driver.find_element(By.NAME,"password")
        #making sure that password input box is clear
        pwd.clear()
        time.sleep(1)
        pwd.send_keys(password)
        #locating login button
        button = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR,"button[type = 'submit']")))
        button.submit()
        #Save Your Login Info? : Not Now
        pop1 = wait.until(EC.presence_of_element_located((By.XPATH,'//div[contains(text(),"Not Now")]')))
        pop1.click()
        #notification Turn on or Not
        pop2 = wait.until(EC.presence_of_element_located((By.XPATH,"//button[contains(@class,'_a9--')]")))
        pop2.click()
        
    except TimeoutException :
        print ("Something went wrong! Try Again")
```

```python
#call the function to login to your instagram handle
#change the login crediential above if you want to login in some different instagram handle
LogIn(username,password)
```

# Problem 2 - Type for "food" in search bar

```python
#this function lets you search for accounts, and it returns the list of all accounts.
def search(value):
    try:
        #locating the search box and cliking on it
        search_box = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR,'svg[aria-label="Search"]')))
        search_box.click()
        #sending the value in the search box
        sending_value = driver.find_element(By.CSS_SELECTOR,'input[aria-label = "Search input"]')
        sending_value.clear()
        sending_value.send_keys(value)
        # waiting till all search is done.
        time.sleep(2)
        #locating and extracting all the search handles. 
        handling_names = wait.until(EC.presence_of_all_elements_located((By.XPATH,"//a[contains(@class,'x1i10hfl')]")))
        names = []
        for i in handling_names:
            try:
                if i.text and i.text[0] != "#":
                    names.append(i.text.split("\n")[0])
            except StaleElementReferenceException:
                 handling_names = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR,"div[role = 'none']")))        
        #clearing the search bar
        driver.find_element(By.CSS_SELECTOR,'div[aria-label = "Clear the search box"]').click()
        #clicking on the search bar again so that it can close.
        search_box = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR,'svg[aria-label="Search"]')))
        search_box.click()
        return names
    except TimeoutException:
        print("No Search Found!")
```

```python
#call the function to get all the searched result
#you can change the value to fetch different accounts
value = 'food'
names = search(value)
for i in names:
    print(i)
```

# Problem 3 - Searching and Opening a profile

```python
#this function lets you search a profile and open it
def search_opening_profiles(value):
    driver.refresh()
    time.sleep(1)
    try:
        #notification Turn On or Not
        pop2 = driver.find_element(By.XPATH,"//button[contains(@class,'_a9--')]")
        pop2.sleep(1)
    except ElementClickInterceptedException:
        pass
    except TimeoutException:
        pass
    except NoSuchElementException:
        pass
    try:
        #Locating the search box and clicking on it
        search_box = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR,'svg[aria-label="Search"]')))
        search_box.click()
        time.sleep(1)
        #sending the value in the search box
        sending_value = driver.find_element(By.CSS_SELECTOR,'input[aria-label = "Search input"]')
        sending_value.clear()
        sending_value.send_keys(value)
        time.sleep(2)
        #waiting till the searched is done
        wait.until(EC.presence_of_element_located((By.XPATH,"//div[contains(@class,'x6s0dn4')]/div/a")))
        time.sleep(1)
        #retry mechanism for StaleElementReferenceException
        for i in range(3):
            try:
                #finding the searched account and clicking on it.
                account = wait.until(EC.presence_of_element_located((By.XPATH,"//div[contains(@class,'x6s0dn4')]/div/a")))
                account.click()
                break
            except StaleElementReferenceException:
                continue
        #after done searching it will remover the search bar and gives you the full page of that Instagram handle.
        try:
            wait.until(EC.presence_of_element_located((By.XPATH,"//div[contains(@class,'_aawn')]")))
            wait.until(EC.presence_of_element_located((By.XPATH,'//div[contains(@class,"_aa_y")]'))).click()
            time.sleep(2)
        except NoSuchElementException:
            pass
    except ElementClickInterceptedException:
        close = driver.find_element(By.CSS_SELECTOR,"svg[aria-label='Close']")
        close.click()
        print("Try Again")
        print("Something Went Wrong!")
    except TimeoutException:
        print("No Search Found!")
```

```python
#call the function to open the instagram account of your choice
#change the value if you want to search any other account
value = "So Delhi"
search_opening_profiles(value)
```

# Follow / Unfollow

Follow/Unfollow given handle -                                                                                                   
1.Open the Instagram Handle of “So Delhi”                                                                                       
2.Start following it. Print a message if you are already following                                                               
3.After following, unfollow the instagram handle. Print a message if you have already unfollowed.

```python
def follow():
    try:
        #locating the follow button
        follow = wait.until(EC.presence_of_element_located((By.XPATH,"//div[contains(@class,'_aacl')]")))
        #if the follow button text is following it will satisfythis condition
        if follow.text == "Following":
            print("You already following the account")
        #else it will follow the account by clicking on follow button
        else:
            follow.click()
    except TimeoutException:
        print("Something Went Wrong")
```

```python
#call the function to follow the account
follow()
```

```python
#this function helps you unfollow accounts and if you already unfollowed that account it tells you that you already unfollowed that account.
def unfollow():
    try:
        unfollow = wait.until(EC.presence_of_element_located((By.XPATH,"//div[contains(@class,'_aacl')]")))
        #if the unfollow button text is follow it will satisfy this condition
        if unfollow.text == "Follow":
            print("You already unfollowed the account")
        else:
            #else it will unfollow the account by clicking on unfollow button.
            unfollow.click()
            time.sleep(1)
            pop_up_of_unfollow = wait.until(EC.presence_of_all_elements_located((By.XPATH,'//div[contains(@class,"x7r02ix")]//span[contains(text(),"Unfollow")]')))
            pop_up_of_unfollow[0].click()
    except TimeoutException:
        print("Something Went Wrong!")
```

```python
#call the function to unfollow the account
unfollow()
```

# Problem 5 - Like/Unlike Post

Like/Unlike posts                                                                                                               
1.Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.                                     
2.Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.

```python
#this function helps you in liking post of any instagram handle and also tells you if you haved already liked the post or Not.
def like_Post():
    try:
        for i in range(2):
        #rendering post with the help of scrollTo
            driver.execute_script('window.scrollBy(0, 6000);')
            time.sleep(3)
        #going upwards using scrollTo
        driver.execute_script('window.scrollBy(0, -9000);')
        time.sleep(1)
        #locating posts
        posts = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME,"_aagw")))
        #looping over posts and getting the required info.
        for i in range(30):
            posts[i].click()
            time.sleep(2)
            #locating the like button
            like = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_aamw')))
            #converting it into parse Tree
            data = BeautifulSoup(like.get_attribute("innerHTML"),"html.parser").title
            #given condition for liked post
            if data.text == "Like":
                like.click()   
                time.sleep(2)  
            else:
                print("You Have Already Liked The Post" + ":" , i+1)
                time.sleep(2)    
            #closing the current post window
            close = driver.find_element(By.CSS_SELECTOR,"svg[aria-label='Close']")
            close.click()
            time.sleep(2)
    except TimeoutException:
        print("Something Went Wrong!")
```

```python
#opening the instagram handle of 'dilsefoodie'
value = "dilsefoodie"
search_opening_profiles(value)
```

```python
#calling the function to like the post
like_Post()
```

```python
#this function helps you in unliking post of any instagram handle and also tells you if you haved already unliked the post or Not.
def unlike_Post():
    try:
        for i in range(2):
        #rendering post with the help of scrollTo
            driver.execute_script('window.scrollBy(0, 6000);')
            time.sleep(3)
        #going upwards using scrollTo
        driver.execute_script('window.scrollBy(0, -6000);')
        time.sleep(1)
        #locating post
        posts = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME,"_aagw")))
        #looping over posts and getting the required info.
        for i in range(30):
            posts[i].click()
            time.sleep(2)
            #locating the unlike button
            unlike = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_aamw')))
            #converting it into parse Tree
            data = BeautifulSoup(unlike.get_attribute("innerHTML"),"html.parser").title
            #given condition for unliked post
            if data.text == "Unlike":
                unlike.click()   
                time.sleep(2)  
            else:
                print("You Have Already Unliked The Post" + ":" , i+1)
                time.sleep(2)   
            #closing the current post window 
            close = driver.find_element(By.CSS_SELECTOR,"svg[aria-label='Close']")
            close.click()
            time.sleep(2)
    except TimeoutException:
        print("Something Went Wrong!")
    except ElementClickInterceptedException:
        close = driver.find_element(By.CSS_SELECTOR,"svg[aria-label='Close']")
        close.click()
        print("Try Again")
```

# Problem 6 - Extract the List of Followers

Extract list of followers                                                                                                     
1.Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’.                                             
2.Now print all the followers of “foodtalkindia” that you are following but those who don’t follow you.

```python
#this functions helps you to extract first 500 followers from Instagram Handles.
def followers_Extractor(value):
    try:
        #locating and opening the followers section by clicking on it.
        time.sleep(3)
        followers_button = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME,"_ac2a")))[1]
        followers_button.click()
        time.sleep(2)
        #getting a list of followers or we can say a frame so that it can help me scrollBy.
        frame = wait.until(EC.presence_of_element_located((By.XPATH, '(//span[contains(@class,"_aacl")])[last()]')))
        #adding a condition for foodtalkindia so that it can only iterate for 2 times.
        n = 90
        if value == "foodtalkindia":
            n = 2
        #loop for scrolling the list to get all the followers
        for i in range(n):
            time.sleep(2)
            driver.execute_script("arguments[0].scrollIntoView(true);",frame)
            frame = wait.until(EC.presence_of_element_located((By.XPATH, '(//span[contains(@class,"_aacl")])[last()-1]')))
        #appending all the followers in names list.
        names = []
        #locating all the followers names.
        followers_list = wait.until(EC.presence_of_all_elements_located((By.XPATH,'//span[contains(@class,"_aacl")]')))
        for i in followers_list[:500]:
            names.append(i.text)
        #after all the process closing the followers window.
        driver.find_element(By.CSS_SELECTOR,"svg[aria-label = 'Close']").click()
        return names
    except ElementClickInterceptedException:
        driver.find_element(By.CSS_SELECTOR,"svg[aria-label = 'Close']").click()
        print("Try Again")     
    except TimeoutException:
        print("Something Went Wrong!")
```

```python
#this functions helps you to extract first 500 followings from Instagram Handles.
def followings_Extractor(value):
    try:
        #locating and opening the following section by clicking on it.
        time.sleep(3)
        followings = driver.find_elements(By.XPATH,"//a[contains(@class,'x1i10hfl')]/span/span")[1]
        followings.click()
        time.sleep(2)
        #getting a list of followings or we can say a frame so that it can help me scrollBy.
        frame = wait.until(EC.presence_of_element_located((By.XPATH, '(//span[contains(@class,"_aacl")])[last()]')))
        #adding a condition for foodtalkindia so that it can only iterate for 2 times.
        n = 90
        if value == "foodtalkindia":
            n = 2
        #loop for scrolling the list to get all the followings
        for i in range(n):
            time.sleep(2)
            driver.execute_script("arguments[0].scrollIntoView(true);",frame)
            frame = wait.until(EC.presence_of_element_located((By.XPATH, '(//span[contains(@class,"_aacl")])[last()-1]')))
        #appending all the followings in names list.
        names = []
        #locating all the following names.
        following_list = wait.until(EC.presence_of_all_elements_located((By.XPATH,'//span[contains(@class,"_aacl")]')))
        for i in following_list[:500]:
            names.append(i.text)
        #after all the process closing the following window.
        driver.find_element(By.CSS_SELECTOR,"svg[aria-label = 'Close']").click()
        return names
    except ElementClickInterceptedException:
        driver.find_element(By.CSS_SELECTOR,"svg[aria-label = 'Close']").click()
        print("Try Again")     
    except TimeoutException:
        print("Something Went Wrong!")
```

```python
#this functions helps you to extract mutual followers who does not follow me back.
def print_non_following_followers_of_me():
    try:
        #locating the mutual follower section and clicking on it
        common_followers_window = wait.until(EC.presence_of_element_located((By.CLASS_NAME,"_aaaj"))).click()
        #after that clicking on more followers so that i can scroll
        see_all_followers = wait.until(EC.presence_of_element_located((By.LINK_TEXT,"See all followers")))
        see_all_followers.click()
        #getting a list of followers or a frame to use this for scrolling
        frame = wait.until(EC.presence_of_element_located((By.XPATH, '(//div[contains(@class,"x1dm5mii")])[last()]')))
        #loop for scrolling and loading all the mutual followers
        for i in range(15):
            time.sleep(2)
            driver.execute_script("arguments[0].scrollIntoView(true);",frame)
            frame = wait.until(EC.presence_of_element_located((By.XPATH, '(//div[contains(@class,"x1dm5mii")])[last()]')))
        #after loading all the followers getting all the list of followers
        common_followers_list = wait.until(EC.presence_of_all_elements_located((By.XPATH,"//span[contains(@class,'_aacl')]")))
        #empty list for mutual followers so that i can append all the followers in that.
        mutual_followers = []
        #loop for appending the followers in mutual_followers.
        for i in common_followers_list:
            mutual_followers.append(i.text)
        #closing the mutual followers section
        driver.find_element(By.CSS_SELECTOR,"svg[aria-label = 'Close']").click()
        time.sleep(1)
        #clicking on our profile to open our Instagram handle
        driver.find_element(By.XPATH,"//span[contains(@class,'xnz67gz')]").click()
        #taking out our own username.
        username = wait.until(EC.presence_of_element_located((By.XPATH,"//h2[contains(@class,'x1lliihq')]")))   
        #extracting all the followings of our account by using followings_extractor function.
        my_followings = followings_Extractor("username")
        #extracting all the followers of our account by using followers_extractor function.
        my_followers = followers_Extractor("username")
        #getting all the mutual that i follow from foodtalkindia
        mutuals_that_i_follow = set(mutual_followers).intersection(my_followings)
        #getting all the mutual who don't follow me back.
        mutuals_who_does_not_follow_me = set(mutuals_that_i_follow).difference(my_followers)
        #returning the result
        return mutuals_who_does_not_follow_me
    except ElementClickInterceptedException:  
        print("Try Again Something Went Wrong")    
        driver.find_element(By.CSS_SELECTOR,"svg[aria-label = 'Close']").click()
    except TimeoutException:
        print("Something Went Wrong!")
```

# Problem 7 - Check the story of ‘coding.ninjas’

```python
#this function lets you see the Instagram handle story and if you already seen the story it will return a msg
#that you have already seen the story and if that account has no story it will give an output that user has no story.
def check_Story():
    try:
        #using height attribute of the user's profile to satisfy above conditons.
        height = wait.until(EC.presence_of_element_located((By.CLASS_NAME,"_aarh"))).get_attribute("height")
        #conditions
        if int(height) == 208:
            print("You have already seen the Story")
        elif int(height) == 168:    
            #locating the story section and clicking on it
            story = wait.until(EC.presence_of_element_located((By.XPATH,"//div[contains(@class,'_aarf _aarg')]")))
            story.click()
            #try except for no story condition
            try:
                time.sleep(2)
                driver.find_element(By.CLASS_NAME,"_ac0y")
                print("Viewing the story")
            except NoSuchElementException:
                print("The user has no story")         
    except TimeoutException:
        print("Something Went Wrong!")
```

```python
#calling the function to check the Instagram handle stories
check_Story()
```
